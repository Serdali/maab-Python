```python
import pandas as pd
df = pd.read_csv('tackoverflow_qa.csv')
df.head()
```

```python
#1. Find all questions that were created before 2014
df['creationdate'] = pd.to_datetime(df['creationdate'])
ques_before_2014 = df[df.creationdate.dt.year < 2014]
ques_before_2014
```

```python
#2. Find all questions with a score more than 50
more_50 = df[df.score > 50]
more_50
```

```python
#3. Find all questions with a score between 50 and 100
ques_between_50_100 = df[(df.score >= 50) & (df.score <= 100)]
ques_between_50_100
```

```python
#4. Find all questions answered by Scott Boston
ans_by_scott = df[df.ans_name == 'Scott Boston']
ans_by_scott
```

```python
#5. Find all questions answered by the following 5 users (Andy Hayden, EdChum, Angel, Jeff, Jim)
five_users = ['Andy Hayden', 'EdChum', 'Angel', 'Jeff', 'Jim']
ans_by_five_users = df[df.ans_name.isin(five_users)]
ans_by_five_users
```

```python
#6. Find all questions created between March 2014 and October 2014 that were answered by Unutbu and have score less than 5
ques_unutbu = df[(df['creationdate'] >= '2014-03-01') & (df['creationdate'] <= '2014-10-31') & (df['ans_name'] == 'unutbu') & (df['score'] < 5)]
ques_unutbu
```

```python
#7. Find all questions that have score between 5 and 10 or have a view count of greater than 10,000
ques_gt_10000 = df[(df['score'].between(5,10)) | (df['viewcount'] > 10000)]
ques_gt_10000
```

```python
#8. Find all questions that are not answered by Scott Boston
ans_not_by_scott = df[df.ans_name != 'Scott Boston']
ans_not_by_scott
```

```python
import pandas as pd

titanic_df = pd.read_csv('titanic.csv')
titanic_df.head()
```

```python
#2. Filter Passengers Who Paid More than $100: Create a DataFrame with passengers who paid a fare greater than $100.
paid_more_100 = titanic_df[titanic_df.Fare > 100]
paid_more_100.head()
```

```python
#4. Filter Passengers Embarked from 'C' and Paid More Than $50
embarked_c_50 = titanic_df[(titanic_df['Embarked'] == 'C') & titanic_df['Fare'] > 50]
embarked_c_50

embarked_c_fare_gt_50 = df[(df['Embarked'] == 'C') & (df['Fare'] > 50)]
```

```python
#5. Select Passengers with Siblings or Spouses and Parents or Children
sibsp_and_parch = titanic_df[(titanic_df['SibSp'] > 0) & (titanic_df['Parch'] > 0)]
sibsp_and_parch
```

```python
#6. Filter Passengers Aged 15 or Younger Who Didn't Survive
age_le_15_not_survived = titanic_df[(titanic_df['Age'] <= 15) & (titanic_df['Survived'] == 0)]
age_le_15_not_survived.head()
```

```python
#7. Select Passengers with Cabins and Fare Greater Than $200
cabin_and_fare_gt_200 = titanic_df[titanic_df['Cabin'].notna() & (titanic_df['Fare'] > 200)]
cabin_and_fare_gt_200.head()
```
